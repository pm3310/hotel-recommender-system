# Data exploration

The [Kaggle Expedia Hotel Recommendations](https://www.kaggle.com/c/expedia-hotel-recommendations) dataset is used to train a Hotel Recommender for email marketing.

The dataset is not used to predict a hotel cluster id based on interactions with the website but to help travelers discover new hotels.

# Exploring the Data

```python
import pandas as pd

train = pd.read_csv('../data/raw/train.csv', nrows=18500000)
```

```python
train.head()
```

```python
train.info()
```

```python
grouped_hotel_clusters = train[train['is_booking'] == 1].groupby('user_id')['hotel_cluster'].apply(list).reset_index(
    name='hotel_clusters_list'
)                                                                                                
```

```python
grouped_hotel_clusters
```

```python
filtered_grouped_hotel_clusters = grouped_hotel_clusters[grouped_hotel_clusters['hotel_clusters_list'].map(len) > 4]
```

```python
filtered_grouped_hotel_clusters
```

```python
sentences = [[str(_item) for _item in _list] for _list in filtered_grouped_hotel_clusters['hotel_clusters_list']]
```
